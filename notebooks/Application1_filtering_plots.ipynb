This notebook is to produce the filtering plots in application I

```python
import warnings   
warnings.simplefilter("ignore") 
import matplotlib as mpl
import matplotlib.pyplot as plt
import pylab
%matplotlib notebook

import numpy as np
import pandas as pd

from scipy.stats import sem
import pickle
```

## CFS

```python
with open("../results/cfs_filtering_updated.pkl", "rb") as f:
    filtlev, sig_bh_cfs, sig_ds_cfs = pickle.load(f)
```

```python
fig_cs = plt.figure()
plt.plot(filtlev, sig_ds_cfs, linestyle='--', marker='o', markeredgewidth=0.0, color = 'red', label = 'DS')
plt.plot(filtlev, sig_bh_cfs, linestyle='--', marker='o', markeredgewidth=0.0, color = 'blue', label = 'BH')
pylab.legend(loc ='upper right', frameon=False)
plt.xlim([-20, 510])
plt.ylim([-2, 70])
plt.xlabel('Filter level', fontsize=20)
plt.ylabel('Bacterial OTUs detected', fontsize=20)
plt.tick_params(labelsize=15)
plt.tight_layout()
fig_cs.savefig('../plots/cfs_filtering.png')
```

```python
# zoom in
fig_cs_zoom = plt.figure()
plt.plot(filtlev[0:20], sig_ds_cfs[0:20], linestyle='--', marker='o', markeredgewidth=0.0,
         color = 'red', label = 'DS')
plt.plot(filtlev[0:20], sig_bh_cfs[0:20], linestyle='--', marker='o', markeredgewidth=0.0,
         color = 'blue', label = 'BH')
plt.xlim([-5, 110])
plt.ylim([-2, 70])
plt.xlabel('Filter level', fontsize=20)
plt.ylabel('Bacterial OTUs detected', fontsize=20)
plt.tick_params(labelsize=15)
plt.tight_layout()
fig_cs_zoom.savefig('../plots/cfs_zoom_filtering.png', bbox_to_inches='tight')
```
