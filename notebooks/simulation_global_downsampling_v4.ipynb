```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
import pylab
%matplotlib notebook
```

## CS

```python
with open("../results_global/results_downsampling/simulation_cs_norm_downSample_fl1000_B100k.pkl", "rb") as f:
    filtlev, B, sample_range, numbact, p_bh, p_fbh, p_ds, err_bh, err_fbh, err_ds = pickle.load(f)
```

```python
print(sample_range)
print(B, filtlev, numbact)
```

```python
print(np.nanmax(err_bh), np.nanmax(err_fbh), np.nanmax(err_ds))
```

```python
# filtlev = 1000
fig_cs_ds = plt.figure()
# plt.axhline(0.1, linestyle='--', color = 'green', markeredgewidth=0.0, label = 'nominal level')
# plt.plot(sample_range, p_ds, linestyle='--', marker='o', color = 'red', markeredgewidth=0.0, label = 'DS')
# plt.plot(sample_range, p_fbh, linestyle='--', marker='o', color = 'orange', markeredgewidth=0.0, label = 'FBH')
# plt.plot(sample_range, p_bh, linestyle='--', marker='o', color = 'blue', markeredgewidth=0.0, label = 'BH')
# pylab.legend(loc ='best', frameon=False)
#plt.suptitle('CS data simulation (B=100000, FilterLev=1000, SE=0.0008)')
plt.axhline(0.1, linestyle='--', color = 'green', markeredgewidth=0.0)
plt.plot(sample_range, p_ds, linestyle='--', marker='o', color = 'red', markeredgewidth=0.0)
plt.plot(sample_range, p_fbh, linestyle='--', marker='o', color = 'orange', markeredgewidth=0.0)
plt.plot(sample_range, p_bh, linestyle='--', marker='o', color = 'blue', markeredgewidth=0.0)
plt.xlabel('Samples per group', fontsize=20)
plt.ylabel('FDR/FWER', fontsize=20)
plt.tick_params(labeltop='off', labelright='off')
plt.tick_params(labelsize=15)
plt.tight_layout()
fig_cs_ds.savefig('plots/plots_v3/cs_simulation_global_ds.png')
```

## CFS
