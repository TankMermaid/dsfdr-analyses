```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.gridspec as gridspec
import pandas as pd
import numpy as np
from scipy.stats import kruskal
%matplotlib inline
```

```python
fig2a=mpimg.imread('../figures/simulation1_sig.png')
fig2b=mpimg.imread('../figures/simulation1_fdr.png')
fig2c=mpimg.imread('../figures/simulation2_sig.png')
fig2d=mpimg.imread('../figures/simulation2_fdr.png')
```

```python
fig2 = plt.figure(figsize=(14, 10))
G = gridspec.GridSpec(2, 2)

axes_1 = plt.subplot(G[0, 0])
plt.imshow(fig2b)
plt.axis('off')
plt.text(0.5, 0.5, 'a', ha='center', va='center', size=24, alpha=1)

axes_2 = plt.subplot(G[0, 1])
plt.imshow(fig2a)
plt.axis('off')
plt.text(0.5, 0.5, 'b', ha='center', va='center', size=24, alpha=1)


axes_3 = plt.subplot(G[1, 0])
plt.imshow(fig2d)
plt.axis('off')
plt.text(0.5, 0.5, 'c', ha='center', va='center', size=24, alpha=1)

axes_4 = plt.subplot(G[1, 1])
plt.imshow(fig2c)
plt.axis('off')
plt.text(0.5, 0.5, 'd', ha='center', va='center', size=24, alpha=1)


plt.subplots_adjust(wspace=-0.3, hspace=None)
plt.tight_layout()
plt.show()

fig2.savefig('../figures/figure2.pdf')
```

```python
fig3a=mpimg.imread('../figures/cfs_filtering.png')
fig3b=mpimg.imread('../figures/mlt_filtering.png')
fig3c=mpimg.imread('../figures/cfs_heatmap.png')
fig3d=mpimg.imread('../figures/mlt_heatmap.png')
fig3e=mpimg.imread('../figures/agp_filtering.png')
fig3f=mpimg.imread('../figures/aga_filtering.png')
```

```python
fig3 = plt.figure(figsize=(15, 15))
G = gridspec.GridSpec(3, 2)

axes_1 = plt.subplot(G[0, 0])
plt.imshow(fig3a)
plt.axis('off')
plt.text(0.5, 0.5, 'a', ha='center', va='center', size=24, alpha=1)
axes_1.set_title('Chronic Fatigue Syndrome', fontsize=15)

axes_2 = plt.subplot(G[0, 1])
plt.imshow(fig3b)
plt.axis('off')
plt.text(0.5, 0.5, 'b', ha='center', va='center', size=24, alpha=1)
axes_2.set_title('Mice Lacking Toll-like Receptor 5', fontsize=15)


axes_3 = plt.subplot(G[1, 0])
plt.imshow(fig3c)
plt.axis('off')
plt.text(0.5, 0.5, 'c', ha='center', va='center', size=24, alpha=1)
axes_3.set_title('Chronic Fatigue Syndrome', fontsize=15)

axes_4 = plt.subplot(G[1, 1])
plt.imshow(fig3d)
plt.axis('off')
plt.text(0.5, 0.5, 'd', ha='center', va='center', size=24, alpha=1)
axes_4.set_title('Mice Lacking Toll-like Receptor 5', fontsize=15)

axes_5 = plt.subplot(G[-1, 0])
plt.imshow(fig3e)
plt.axis('off')
plt.text(0.5, 0.5, 'e', ha='center', va='center', size=24, alpha=1)
axes_5.set_title('American Gut Plant Number', fontsize=15)

axes_6 = plt.subplot(G[-1, 1])
plt.imshow(fig3f)
plt.axis('off')
plt.text(0.5, 0.5, 'f', ha='center', va='center', size=24, alpha=1)
axes_6.set_title('American Gut Antibiotic History', fontsize=15)


plt.subplots_adjust(wspace=-0.5, hspace=None)
plt.tight_layout()
plt.show()

fig3.savefig('../figures/figure5.pdf')
```
