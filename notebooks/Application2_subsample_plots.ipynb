This notebook is to produce the subsampling figures in application II.

```python
import warnings   
warnings.simplefilter("ignore") 
import matplotlib as mpl
import matplotlib.pyplot as plt
import pylab
%matplotlib notebook
%matplotlib inline

import numpy as np

from scipy.stats import sem
import pickle
```

## DME

```python
# input DME subsampling results
with open("../results/dme_subsample_v2.pkl", "rb") as f:
    ss_dme, otu_ds_dme, otu_bh_dme, err_ds_dme,err_bh_dme = pickle.load(f)
```

```python
# plot OTUs detected vs subsample size
fig_dme = plt.figure()
plt.errorbar(ss_dme, otu_ds_dme, yerr=err_ds_dme, lw=2, color = 'red', label = 'DS')
plt.errorbar(ss_dme, otu_bh_dme, yerr=err_bh_dme, lw=2, color = 'blue', label = 'BH')
pylab.legend(loc ='lower right', frameon=False)
plt.xlabel('Samples per group', fontsize=20)
plt.ylabel('Bacterial OTUs detected', fontsize=20)

plt.plot([7.5, 16], [np.max(otu_bh_dme), np.max(otu_bh_dme)], 'k-', lw=2, linestyle='--', 
         color = 'orange')
plt.xticks(np.arange(2, 18, 2))
plt.tick_params(labeltop='off', labelright='off')
plt.tick_params(labelsize=15)
plt.tight_layout()
fig_dme.savefig('../plots/dme_subsample.png')
```

```python
# statistics
print(np.max(otu_bh_dme))
(8-16)/16*100
```
