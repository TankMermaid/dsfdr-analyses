This notebook is to generate Figure 3 and 4 in the manuscript for data driven simulations.

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.gridspec as gridspec
import pandas as pd
import numpy as np
from scipy.stats import kruskal
%matplotlib inline
```

## Figure 3

```python
fig3a=mpimg.imread('../plots/dibd_simulation_global_nulls.png')
fig3b=mpimg.imread('../plots/cs_simulation_global_nulls.png')
fig3c=mpimg.imread('../plots/dibd_simulation_global_fl.png')
fig3d=mpimg.imread('../plots/cs_simulation_global_fl.png')
fig3e=mpimg.imread('../plots/dibd_simulation_global_ds.png')
fig3f=mpimg.imread('../plots/cs_simulation_global_ds.png')
```

```python
fig3 = plt.figure(figsize=(15, 15))
G = gridspec.GridSpec(3, 2)

axes_1 = plt.subplot(G[0, 0])
plt.imshow(fig3a)
plt.axis('off')
plt.text(0.5, 0.5, 'a', ha='center', va='center', size=24, alpha=1)
axes_1.set_title('DIBD simulation with no signals', fontsize=15)

axes_2 = plt.subplot(G[0, 1])
plt.imshow(fig3b)
plt.axis('off')
plt.text(0.5, 0.5, 'b', ha='center', va='center', size=24, alpha=1)
axes_2.set_title('CS simulation with no signals', fontsize=15)


axes_3 = plt.subplot(G[1, 0])
plt.imshow(fig3c)
plt.axis('off')
plt.text(0.5, 0.5, 'c', ha='center', va='center', size=24, alpha=1)


axes_4 = plt.subplot(G[1, 1])
plt.imshow(fig3d)
plt.axis('off')
plt.text(0.5, 0.5, 'd', ha='center', va='center', size=24, alpha=1)

axes_5 = plt.subplot(G[-1, 0])
plt.imshow(fig3e)
plt.axis('off')
plt.text(0.5, 0.5, 'e', ha='center', va='center', size=24, alpha=1)

axes_6 = plt.subplot(G[-1, 1])
plt.imshow(fig3f)
plt.axis('off')
plt.text(0.5, 0.5, 'f', ha='center', va='center', size=24, alpha=1)


plt.subplots_adjust(wspace=-0.5, hspace=None)
plt.tight_layout()
plt.show()

fig3.savefig('../figures/figure3.pdf', dpi=300)
```
